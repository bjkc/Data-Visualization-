# Proportions

```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head()
```

```python
df.groupby('workout_type').workout_type.agg('count')
```

```python
plt.pie(x=df.groupby('workout_type').workout_type.agg('count'),labels = df.groupby('workout_type').workout_type.agg('count').index, autopct='%.0f%%' )
plt.show()
```

Here we can see blue color for both cricket and walking. 

```python
import matplotlib.pyplot as plt
plt.figure()
values = [93, 96, 80, 91,85, 81, 94, 90, 98, 94, 98]
labels= ['Cricket','Freestyle','Indoor Cycling','Open Water','Outdoor Cycling',    
'Outdoor Running',   
'Pool Swimming',      
'Trail Run' ,        
'Treadmill' ,         
'Trekking'  ,         
'Walking' ]  
plt.pie(values, labels=labels, autopct='%.2f')
plt.show()
```

I did change the percentage to number however, the color is still same for **Cricket** and **Walking**

```python
import numpy as np
from numpy import multiply
import pylab as pl
from matplotlib import collections  as mc
fracs = [93, 96, 80, 91,85, 81, 94, 90, 98, 94, 98]
colors = sns.color_palette('bright')[:8]
labels= ['Cricket','Freestyle','Indoor Cycling','Open Water','Outdoor Cycling',    
'Outdoor Running',   
'Pool Swimming',      
'Trail Run' ,        
'Treadmill' ,         
'Trekking'  ,         
'Walking' ]          
total = sum(fracs)
plt.pie(fracs, labels=labels,
    autopct =lambda p: '{:.2f}'.format(p * total / 100),
    shadow=False, startangle=90,colors= colors)
```

I think the color is kind of misleading. 

If I put that same data into a bar chart.

```python
ax = sns.barplot( x= df.groupby('workout_type').workout_type.agg('count').index,
                 y = df.groupby('workout_type').workout_type.agg('count'), palette = 'bright')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
ax.set_title('Bar-Chart')

plt.show()
```

## Mosaic Plots

```python
from statsmodels.graphics.mosaicplot import mosaic
df2 = df[(df.activity_day == '2022-01-01')|(df.activity_day == '2022-01-02')|(df.activity_day == '2022-01-03')]
ax = mosaic(df2, [ 'workout_type','activity_day'], axes_label= True, label_rotation=90, title ='Graph', horizontal =True, gap=0.025 )
plt.show()
```

```python
df2.groupby(['workout_type','activity_day']).workout_type.agg('count')
```

```python
ax = sns.barplot( x= df.groupby('workout_type').workout_type.agg('count').index, y = df.groupby('workout_type').workout_type.agg('count'))
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
ax.annotate("Crazy Fan",
            xy = [0,90],
            xytext = [10,110],
            arrowprops = dict(facecolor = 'red', width = 1, headwidth = 15, headlength = 10))
ax.set_title('Annotated Graph')
ax.set_ylabel('Frequency')

plt.show()
```

```python
ax = sns.barplot( x= df.groupby('workout_type').workout_type.agg('count').index, y = df.groupby('workout_type').workout_type.agg('count'))
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
ax.annotate("Guess! How much?",
            xy = [5,0],
            xytext = [10,110],
            arrowprops = dict(facecolor = 'red', width = 1, headwidth = 15, headlength = 10))
ax.set_title('Annotated Graph')
ax.set_ylabel('Frequency')

plt.show()
```

Mosaic plot helps us to compare groups. It is also know as two stacked bar chart. It is a good way to compare categorical variable. It can show both information that can be shown using **pie-chart** and **bar-chart**. **Pie-chart** is well know graph for categorical variable analysis at the same time it needs to have some numerical value. And for the **bar-chart** we can use it to show the visual of catagorical variable. However, as we can see in the **Moasic plot** we can visual both. Even though it has some disadvangates like comparing the rectange can sometimes be difficult, it is still a great visual over two.
