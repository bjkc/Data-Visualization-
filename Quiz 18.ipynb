```python
import pandas as pa

df = pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv")
df.head()
```

```python
pa.get_dummies(df.neighbourhood_group).head()
```

```python
df1 = pa.concat([df,pa.get_dummies(df.neighbourhood_group)],axis = 1)

df1 = df1.drop('neighbourhood_group', axis = 1)

df1
```

Just like an example shown in the class, I have used dummies function. We can clearly see it did a great job in converting into the numeric form. It has extra rows but I do not see a problem with that. It can also be done differently like it is done below making a loop. 

```python
neighbour = sorted(df.neighbourhood_group.unique())

neighbour
```

```python
eddict = {}

for i in range(5):
  eddict[neighbour[i]] = i

eddict
```

```python
df.groupby('neighbourhood_group').host_id.agg('count')
```

```python
df.neighbourhood_group.map(eddict)
```

```python
df.neighbourhood_group=df.neighbourhood_group.map(eddict)
df.head()
```
