##Finding NaN's


```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')

df.head()
```

```python
df.isnull().sum(axis = 0)
```

There are null values in only few columns.

```python
df[~df.isnull().any(axis = 1)]
```

```python
df[df.isnull().sum(axis = 1) <= 4]
```

```python
df[~df.last_review.isnull()]
```

Another method for filling in NaN shown below. 

```python
df.fillna(value = 0)
```

```python
df.last_review.fillna(value = 0)
```

```python
df.last_review[~df.last_review.isnull()]
```

```python
df.last_review.fillna(method= 'pad')
```

I think the second table has increased the length to 48895 from 38843. 

```python
df.last_review.fillna(method= 'bfill')
```

```python
df.reviews_per_month.fillna(method= 'bfill')
```

```python
df.reviews_per_month.transform(lambda x: x.fillna(value = x.mode()[0]))
```

It looks like it replaced NaN with 0.02. Is it normal? Let's see what it does to last_reviews. 

```python
df.last_review.transform(lambda x: x.fillna(value = x.mode()[0]))
```

In this case NaN is replaced by 2019-06-23.
