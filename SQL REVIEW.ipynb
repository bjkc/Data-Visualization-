```python
%%bigquery --project data-base-1-338717
SELECT C.country_code, C.country_name , I.country_area 
FROM `bigquery-public-data.census_bureau_international.country_names_area` as I
JOIN`bigquery-public-data.census_bureau_international.midyear_population` as C 
ON I.country_code=C.country_code
Where C.country_code like 'S%'
GROUP BY C.country_code, C.country_name, I.country_area
ORDER BY country_area DESC
```
