I am importing the necessary Library that are pandas, seaborn and matplot. I have used **df.head(30)** because it prints 30 rows. 

```python
df.info()
```

The above table can be very handy when it comes to visualization. It provides us with the information of a data type we are going to be dealing with. That is why I have used **info()** command. Here, we can see some of the data are integers, some of them are objects and some of them are floats.

As we can see how messy above graph is. I can not get any information from it. It shows me the histogram about the **Goals**. However, I would like to look particular in the **Goals.**

```python
#ax = plt.hist(df.Goals, bins = 15, density = True, histtype='bar', align= 'mid' , orientation ='vertical',rwidth = 5 ,log = False, color = None, Label = [7]) could not put the title
ax = sns.histplot(data = df, x = 'Goals', y= None)
ax.set_title("Distribution of Goals")
plt.show()
```

As we can see it is clear and easy to understand this histogram. Histogram helps us to understand the distribution of a data over a continuous interval. It is showing a numeric data in a range. From the above graph we can tell it is right-skewed distribution. It is true that it is not that easy to score too many goals. Players not scoring is higher than player who score 20 goals. 
[link](**`https://github.com/khatriprajwol/Data-Visualization-/blob/main/Histogram.ipynb`**)

```python
ax = sns.histplot(data = df, x = 'Goals', y= None, hue='nationality', )
ax.set_title("Distribution of Goals by nationality")
plt.show()
```

This histogram shows the goal distribution by nationality.
