```python
%%bigquery --project data-base-1-338717 
SELECT  taxi_id,AVG(trip_seconds) as avg_trip_seconds,
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
GROUP BY taxi_id
ORDER BY avg_trip_seconds DESC
LIMIT 10
```

**https://github.com/nurfnick/Data_Viz/blob/main/12_GatheringConclusion.ipynb. I attest that the resources above were the only ones utilized in completing the exam and the work included is my own and no one else from the course.**

In all the tables I have used `LIMIT 10` because not having it was time consuming lots of rows. Every `taxi_id` has its `trips_seconds` and using `AVG` function gave me each `taxi_id's`average trips. Here `43195` is the highest average.

```python
%%bigquery --project data-base-1-338717 
SELECT taxi_id, trip_seconds,
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
WHERE (trip_seconds != 0)
ORDER BY trip_seconds DESC
LIMIT 10
```

Not using `!=0` gave me `0` in my table that is why I have used the `WHERE` clauses to remove `0. 

```python
%%bigquery --project data-base-1-338717
WITH t as (
SELECT   taxi_id,AVG(trip_seconds) as avg_trip_seconds
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
GROUP BY taxi_id
)
SELECT t1.taxi_id, COUNT(*) as above_avg_time
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips` t1 JOIN t  ON t1.taxi_id = t.taxi_id 
WHERE (t1.trip_seconds > t.avg_trip_seconds)
GROUP BY taxi_id
ORDER BY above_avg_time DESC
limit 10
```
